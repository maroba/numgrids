# The Hydrogen atom

```python
from numgrids import *
import numpy as np
```

```python
cut_off = 1.E-2

grid = SphericalGrid(
    Axis.of_type(AxisType.CHEBYSHEV, 60, cut_off, 50, name="r"),  # radial axis
    Axis.of_type(AxisType.CHEBYSHEV, 30, cut_off, np.pi - cut_off, name=r"\theta"), # polar axis
    Axis.of_type(AxisType.EQUIDISTANT_PERIODIC, 50, cut_off, 2*np.pi, name=r"\varphi"),  # azimuthal axis
)
```

```python
R, Theta, Phi = grid.meshed_coords
```

```python
H = lambda psi: -0.5 * grid.laplacian(psi) - 2 / R * psi
```

```python
psi0 = np.exp(-R)
```
